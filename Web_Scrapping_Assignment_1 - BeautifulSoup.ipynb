# 1.Write a python program to display all the header tags from 'https://en.wikipedia.org/wiki/Main_Page'Â¶

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

Since the response is 200, I can scrap the data

```python
soup=BeautifulSoup(page.content)
soup
```

```python
Header_tags=[]
header=soup.find_all(["h1","h2","h3","h4","h5","h6"])
for i in header:
    Header_tags.append(i.text.replace("\n",""))
    
Header_tags
```

# Making the DataFrame

```python
import pandas as pd
wiki=pd.DataFrame({})
wiki['Header tags']=Header_tags
wiki
```

# 9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

# Scraping the restaurant name

```python
restaurant_one=soup.find('div',class_='restnt-info cursor')
restaurant_one
```

```python
restaurant_one.text
```

```python
Restaurants=[]
for i in soup.find_all('div',class_='restnt-info cursor'):
    Restaurants.append(i.text)
Restaurants
```

# Scraping the cuisine

```python
cui=soup.find('div',class_='detail-info')
cui
```

```python
cui.text
```

```python
cui.text.split()[6]
```

```python
Cuisine=[]
for i in soup.find_all('div',class_='detail-info'):
    Cuisine.append(i.text.split()[6])
Cuisine
```

# Scraping the location

```python
loc=soup.find('div','restnt-loc ellipsis')
loc
```

```python
loc.text
```

```python
Location=[]
for i in soup.find_all('div','restnt-loc ellipsis'):
    Location.append(i.text)
Location
```

```python
rtg=soup.find('div',class_='restnt-rating rating-4')
rtg
```

```python
rtg.text
```

```python
Ratings=[]
for i in soup.find_all('div',class_='restnt-rating rating-4'):
    Ratings.append(i.text)
Ratings
```

# Scraping the image URL

```python
img_url=soup.find_all('img',class_='no-img')
img_url
```

```python
image_url=[]

img_url=soup.find_all("img",class_="no-img")
for i in img_url:
    image_url.append(i.get("data-src"))
image_url
```

# Framing the DatasetÂ¶

```python
import pandas as pd
df=pd.DataFrame({})
df['Restaurants']=Restaurants
df['Cuisine']=Cuisine
df['Location']=Location
df['Ratings']=Ratings
df['image_url']=image_url
df
```
